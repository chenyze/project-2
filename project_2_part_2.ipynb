## Dropping features that were omitted from clean_X

```python
test.columns
```

```python
test.columns = test.columns.str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '').str.replace('/', '_').str.replace('.', 'n')
test_col = [col for col in test.columns]
test_col
```

```python
print(clean_train.shape)

train_col = [ col.lower() for col in clean_train.columns]
train_col
```

```python
# we will use this list to track what are the columns that we don't have in the cleaned training set
not_in_train = []
```

```python
for i in test_col:
    if i not in train_col: 
        not_in_train.append(i)
    else:
        continue

not_in_train
```

We can't just drop the columns in not_in_train from the test dataframe - some of the discrepeancies might have arisen from using get_dummies() on the test data set so we'll need to look at those.

```python
# keep_col will store the test column names that are substrings of columns in train.
# for instance we want to keep land_contour because it corresponds to land_contour_hls (a result of get_dummies) in train

keep_col = []

for col in not_in_train:
    count = 0
    for c in train_col:
        if col in c:
            # print(c)
            count = count + 1
            # print("count is:", count)
        else:
            pass
    if count > 0:
        keep_col.append(col)
    else:
        pass
    
keep_col
```

```python
drop_cols = []

for k in not_in_train:
    if k not in keep_col: 
        drop_cols.append(k)
    else:
        continue
        
drop_cols
```

Even then, there are a couple of columns that are essential for feature-engineering to match our train data set so we want to omit these from drop-cols

* 'bsmt_full_bath' - for calculating 'full_bath' and 'total_baths'
* 'bsmt_half_bath' - for calculating 'total_baths'
* 'half_bath' - for calculating 'total_baths'

```python
len(drop_cols)
```

```python
drop_cols.remove('bsmt_full_bath')
drop_cols.remove('bsmt_half_bath')
drop_cols.remove('half_bath')

len(drop_cols)
```

```python
# dropping the unnecessary columns
print(test.shape)
test.drop(columns=drop_cols, inplace=True )
print(test.shape)
```
